```python
import pandas as pd
import matplotlib.pyplot as plt

# Carregar e Limpar base
df = pd.read_csv("/content/datatran2025.csv", sep=';', encoding='latin1', low_memory=False)

def clean_float(col):
    return pd.to_numeric(
        col.astype(str).str.replace(',', '.', regex=False).str.replace(' ', '', regex=False),
        errors='coerce'
    )

df['km'] = clean_float(df['km'])
df['latitude'] = clean_float(df['latitude'])
df['longitude'] = clean_float(df['longitude'])

df_sp_116 = df[(df['uf'] == 'SP') & (df['br'] == 116)].copy()
df_sp_116['hora'] = df_sp_116['horario'].astype(str).str.slice(0,2)

# Estatísticas
top_mun = df_sp_116['municipio'].value_counts().head(10)
top_tipos = df_sp_116['tipo_acidente'].value_counts().head(10)
horas = df_sp_116['hora'].value_counts().sort_index()

# 1. Municípios
plt.figure(figsize=(10,5))
top_mun.plot(kind='bar')
plt.title("Top 10 Municípios com Mais Sinistros na BR-116/SP (2025)")
plt.xlabel("Município")
plt.ylabel("Quantidade de Sinistros")
plt.xticks(rotation=45, ha="right")
plt.tight_layout()
plt.savefig("/content/grafico_municipios.png")
plt.close()

# 2. Tipos de acidente
plt.figure(figsize=(10,5))
top_tipos.plot(kind='bar')
plt.title("Top 10 Tipos de Acidente na BR-116/SP (2025)")
plt.xlabel("Tipo de Acidente")
plt.ylabel("Quantidade")
plt.xticks(rotation=45, ha="right")
plt.tight_layout()
plt.savefig("/content/grafico_tipos.png")
plt.close()

# 3. Horários
plt.figure(figsize=(10,5))
horas.plot(kind='bar')
plt.title("Distribuição de Sinistros por Hora do Dia – BR-116/SP (2025)")
plt.xlabel("Hora")
plt.ylabel("Quantidade")
plt.xticks(rotation=0)
plt.tight_layout()
plt.savefig("/content/grafico_horas.png")
plt.close()

"/content/grafico_municipios.png", "/content/grafico_tipos.png", "/content/grafico_horas.png"
```

```python
# Preparar dataset final para LSTM

import pandas as pd # Ensure pandas is imported

# Reload the original DataFrame to ensure 'uf' and 'br' columns are present
df = pd.read_csv("/content/datatran2025.csv", sep=';', encoding='latin1', low_memory=False)

def clean_float(col):
    return pd.to_numeric(
        col.astype(str).str.replace(',', '.', regex=False).str.replace(' ', '', regex=False),
        errors='coerce'
    )

df['km'] = clean_float(df['km'])
df['latitude'] = clean_float(df['latitude'])
df['longitude'] = clean_float(df['longitude'])

# Filtro SP + BR116
df = df[(df['uf']=='SP') & (df['br']==116)].copy()

# Corrigir formato Data/Horário
df['data_inversa'] = pd.to_datetime(df['data_inversa'], errors='coerce')

df['horario'] = df['horario'].astype(str).str.slice(0,5)
df['horario'] = pd.to_datetime(df['horario'], format="%H:%M", errors='coerce').dt.time

df['timestamp'] = df.apply(
    lambda x: pd.Timestamp.combine(x['data_inversa'], x['horario'])
    if pd.notnull(x['data_inversa']) and pd.notnull(x['horario']) else pd.NaT,
    axis=1
)

df = df.dropna(subset=['timestamp']).sort_values('timestamp')

# Agregar horário
df_hourly = df.resample("1h", on="timestamp").size().reset_index(name="qtde_sinistros")

output_path = "/content/dataset_lstm_br116_sp_2025.csv"
df_hourly.to_csv(output_path, index=False)

df_hourly.head(), output_path
```

```python
import pandas as pd
import numpy as np
from sklearn.preprocessing import MinMaxScaler

# Carregar dataset tratado
df = pd.read_csv("/content/dataset_lstm_br116_sp_2025.csv", parse_dates=["timestamp"])

# Sort by timestamp
df = df.sort_values("timestamp").reset_index(drop=True)

# 1. Normalização
scaler = MinMaxScaler()
df["qtde_norm"] = scaler.fit_transform(df[["qtde_sinistros"]])

# 2. Criar janelas de tempo (ex.: usar 24h anteriores para prever a hora seguinte)
WINDOW_SIZE = 24

X = []
y = []

values = df["qtde_norm"].values

for i in range(WINDOW_SIZE, len(values)):
    X.append(values[i-WINDOW_SIZE:i])
    y.append(values[i])

X = np.array(X)
y = np.array(y)

# 3. Divisão treino/validação/teste
train_size = int(len(X) * 0.70)
val_size = int(len(X) * 0.15)

X_train, y_train = X[:train_size], y[:train_size]
X_val,   y_val   = X[train_size:train_size+val_size], y[train_size:train_size+val_size]
X_test,  y_test  = X[train_size+val_size:], y[train_size+val_size:]

# Salvar os conjuntos para utilização no modelo LSTM
np.save("/content/X_train.npy", X_train)
np.save("/content/y_train.npy", y_train)
np.save("/content/X_val.npy",   X_val)
np.save("/content/y_val.npy",   y_val)
np.save("/content/X_test.npy",  X_test)
np.save("/content/y_test.npy",  y_test)

"/content/X_train.npy", "/content/X_val.npy", "/content/X_test.npy"
```

```python
# Load dataset
df = pd.read_csv("/content/dataset_lstm_br116_sp_2025.csv", parse_dates=["timestamp"])
df = df.sort_values("timestamp")

# Create time-series plots directory
# 1) Raw hourly series
plt.figure(figsize=(12,5))
plt.plot(df["timestamp"], df["qtde_sinistros"])
plt.title("Série Temporal – Sinistros por Hora (BR-116/SP – 2025)")
plt.xlabel("Tempo")
plt.ylabel("Qtde de Sinistros")
plt.tight_layout()
plt.savefig("/content/serie_temporal_raw.png")
plt.close()

# 2) 24h moving average
df["rolling_24h"] = df["qtde_sinistros"].rolling(window=24, min_periods=1).mean()

plt.figure(figsize=(12,5))
plt.plot(df["timestamp"], df["rolling_24h"])
plt.title("Média Móvel de 24 Horas – Sinistros (BR-116/SP – 2025)")
plt.xlabel("Tempo")
plt.ylabel("Média de Sinistros (24h)")
plt.tight_layout()
plt.savefig("/content/serie_temporal_rolling24h.png")
plt.close()

"/content/serie_temporal_raw.png", "/content/serie_temporal_rolling24h.png"
```
